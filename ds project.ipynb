```python
#found the outliers that might drive the outcome of the model
#finding like users on the base of stars
shaped['full_name']=shaped['owner']+'/'+shaped['repo']
shaped.head(10)
```

```python
counts = shaped['full_name'].value_counts()
shaped = shaped[~shaped['full_name'].isin(counts[counts >40].index)]
shaped.shape
```

```python
# matrix for svd usage
watchers_matrix=pd.crosstab(shaped['id'],shaped['full_name'])
watchers_matrix.columns
```

```python
#applying svd on matrix
from scipy.sparse.linalg import svds
watchers_matrix=watchers_matrix.astype(float)
U,S,V=svds(watchers_matrix,k=12)
```

```python
import numpy as np
A=np.diag(S)
arr=np.dot(np.dot(U,A),V)
print(arr)
pred_df=pd.DataFrame(arr)
```

```python
#finding the top repos recommended to each user
pred_df['id']=shaped['id']
final_dict={}
for row in pred_df.iterrows():
    temp=[]
    final_list=[]
    already_starred=list(shaped[shaped['id']==row[1][-1]].full_name)
    sorted_user_pred=row[1].sort_values(ascending=False)[1:10]
    sorted_user_pred=pd.DataFrame(sorted_user_pred)
    for i in range(9):
        temp.append(sorted_user_pred.iloc[i].name)
    for i in temp:
        if(i not in already_starred):
            final_list.append(i)
    final_dict[row[1][-1]]=final_list
```
