```python
f = open("./here1.txt", "r")
langUserDict = {}
count = 0
for s in f:
    s_split = s.split(" : ")
    langListString = s_split[1]
    langListString = langListString.replace("{", "")
    langListString = langListString.replace("}", "")
    langListString = langListString.replace("\'", "")
    langListString = langListString.replace("\n", "")
    langList = langListString.split(" ")
    for x in langList:
        if x== 'None':
            continue
        if x in langUserDict.keys():
            langUserDict[x] += 1
        else:
            langUserDict[x] = 1
    count += 1
print(langUserDict)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(60,30))
keys = []
values = []
for lang in langUserDict.keys():
    if(langUserDict[lang]>50):
        keys.append(lang)
        values.append(langUserDict[lang])
        
y_pos = np.arange(len(keys))
 
# Create bars
plt.bar(y_pos, values)
 
# Create names on the x-axis
plt.xticks(y_pos, keys, color='orange', rotation = 90)
plt.yticks(color='orange')
plt.rc('xtick', labelsize=40)   
plt.rc('ytick', labelsize=40) 

plt.show()
```
